### **1. Introduction**
The **Human Activity Recognition 70+ (HAR70+) dataset** is a time-series dataset collected from **elderly individuals (aged 70-95 years)** using **wearable sensors** (Axivity AX3 accelerometers) attached to the lower back and thigh.  
The dataset is annotated for **eight different activities**, useful for developing **IoT-based activity recognition models**.

  - A **deep learning-based state change detector** using **LSTM**.
  - A **traditional ML classifier** using **Random Forest**.

## **2. Data Loading & Preprocessing**


```python
# **Import Libraries**
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.utils import to_categorical

# **Set Dataset Directory**
data_dir = r"har70plus"  

# **Load All CSV Files in the Folder**
def load_har70_data(data_dir):
    all_files = [f for f in os.listdir(data_dir) if f.endswith('.csv')]
    data_list = []

    for file in all_files:
        file_path = os.path.join(data_dir, file)
        df = pd.read_csv(file_path)
        df['file_id'] = file  # Track participant
        data_list.append(df)

    return pd.concat(data_list, ignore_index=True)

# **Load Data**
df = load_har70_data(data_dir)

# **Inspect Data**
df.info()
print(df.head())
print(df.tail())
```

```python
# **Train LSTM Model**
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense

# Define LSTM model
model_lstm = Sequential([
    LSTM(64, return_sequences=True, input_shape=(X_lstm.shape[1], X_lstm.shape[2])),
    LSTM(32),
    Dense(2, activation='softmax')  # Binary classification
])

# Compile model
model_lstm.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Train the model
history = model_lstm.fit(X_lstm, y_lstm, epochs=10, batch_size=32, shuffle=False, validation_split=0.2)
```

```python
# **Import Libraries for Evaluation**
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import seaborn as sns

# **Make Predictions on Test Set**
y_pred = model_lstm.predict(X_lstm)  # Predict probabilities
y_pred_classes = np.argmax(y_pred, axis=1)  # Convert to binary (0 or 1)
y_true_classes = np.argmax(y_lstm, axis=1)  # Convert categorical back to labels

# **Print Classification Report**
print("Classification Report:\n", classification_report(y_true_classes, y_pred_classes))

# **Plot Confusion Matrix**
cm = confusion_matrix(y_true_classes, y_pred_classes)
plt.figure(figsize=(6,5))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['No Change', 'Change'], yticklabels=['No Change', 'Change'])
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix for State Transition Detection")
plt.show()
```

```python
# **Plot Training Loss & Accuracy**
plt.figure(figsize=(12,5))

# Loss Curve
plt.subplot(1,2,1)
plt.plot(history.history['loss'], label="Training Loss")
plt.plot(history.history['val_loss'], label="Validation Loss")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.title("Training vs. Validation Loss")
plt.legend()

# Accuracy Curve
plt.subplot(1,2,2)
plt.plot(history.history['accuracy'], label="Training Accuracy")
plt.plot(history.history['val_accuracy'], label="Validation Accuracy")
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.title("Training vs. Validation Accuracy")
plt.legend()

plt.show()
```
